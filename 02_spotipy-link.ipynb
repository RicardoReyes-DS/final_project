```python
# Spotify Artist Info Function

def artists_info(artists_list):
    
    '''
    
    artists_list: SpotifyArtistID List
    
    '''
    
    results=spotify.artists(artists_list)
    
    artist_id=[]
    artist_name=[]
    artist_genres=[]
    artist_followers=[]
    artists_popularity=[]

    for artist in results['artists']:
        artist_id.append(artist['id'])
        artist_name.append(artist['name'])
        artist_genres.append(artist['genres'])
        artist_followers.append(artist['followers']['total'])
        artists_popularity.append(artist['popularity'])
        
    columns=['SpotifyArtistID', 'ArtistName', 'ArtistGenres', 'SpotifyArtistFollowers', 'SpotifyArtistPopularity']
        
    df_artists_info=pd.DataFrame(list(zip(artist_id,
                                          artist_name,
                                          artist_genres,
                                          artist_followers,
                                          artists_popularity)),
                                 columns=columns) 
    
    #print(df_artists_info.info())
    return df_artists_info
```

```python
# Spotify Artists' Albums Function

def artist_albums_df(artists_id):
    '''
    
    INPUT: artists_id: SpotifyArtistID
    OUTPUT: DataFrame with Artist's Albums: 
    (album_id, artist_id, album_name, album_release_date, album_available_markets)
    
    '''
    
    # results = spotify.artist_albums(artists_id, album_type='album')
    results = spotify.artist_albums(artists_id)
    album_id=[]
    artist_id=[]
    album_name=[]
    album_release_date=[]
    album_available_markets=[]

    for album in results['items']:
        album_id.append(album['id'])
        artist_id.append(album['artists'][0]['id'])
        album_name.append(album['name'])
        album_release_date.append(album['release_date'])
        album_available_markets.append(album['available_markets'])

    columns=['SpotifyAlbumID', 'SpotifyArtistID', 'AlbumName', 'SpotifyReleaseDate', 'SpotifyMarkets']

    df_artist_albums=pd.DataFrame(list(zip(album_id,
                                           artist_id,
                                           album_name,
                                           album_release_date,
                                           album_available_markets)), columns=columns)
    
    df_artist_albums['SpotifyReleaseDate'] = pd.to_datetime(df_artist_albums['SpotifyReleaseDate'],
                                                         format='%Y-%m-%d',
                                                         errors='ignore')

    print(df_artist_albums.info())
    return df_artist_albums
```

```python
# Related Artists by Single Artist Function

def related_artists(artist_id):
    
    '''
    
    INPUT: Individual artist_id
    OUTPUT: DataFrame RelatedFromSID, SpotifyArtistID, ArtistName, SpotifyArtistFollowers, SpotifyArtistPopularity
    
    '''
    
    results = spotify.artist_related_artists(artist_id)

    # Extracting Related Artists Info form dictionary

    related_from=[]
    related_artists = []
    related_artists_followers = []
    related_artists_popularity = []
    related_artists_id = []

    for item in results['artists']:
        related_from.append(artist_id)
        related_artists.append(item['name'])
        related_artists_followers.append(item['followers']['total'])
        related_artists_popularity.append(item['popularity'])
        related_artists_id.append(item['id'])

    columns=['RelatedFromSID','SpotifyArtistID','ArtistName','SpotifyArtistFollowers','SpotifyArtistPopularity']

    df_related_artists=pd.DataFrame(list(zip(related_from,
                                             related_artists_id,
                                             related_artists,
                                             related_artists_followers,
                                             related_artists_popularity)),
                                   columns=columns)
    #print(df_related_artists.info())
    return df_related_artists

    # RELATED ARTISTS WILL ALSO BE ANALYZED TO GET TOP TRACKS
```

```python
# Artist Top Tracks

'''
TOP TRACKS MUST BE ADDED TO PLAYLIST
'''

# THIS FUNCTION RECEIVES ARTISTS ID


results=spotify.artist_top_tracks(marcel_dettmann_uri)

top_track_author=[]
artist_top_tracks=[]
track_popularity=[]
track_release_date=[]
track_spotify_id=[]

for track in results['tracks']:
    artist_top_tracks.append(track['name'])
    track_popularity.append(track['popularity'])
    track_release_date.append(track['album']['release_date'])
    track_spotify_id.append(track['id'])
    top_track_author.append(track['artists'][0]['name'])

columns = ['SpotifyTrackID','ArtistsName', 'TrackName', 'TrackPopularity', 'ReleaseDate']
    
df_artist_top_tracks=pd.DataFrame(list(zip(track_spotify_id,
                                           top_track_author,
                                           artist_top_tracks,
                                           track_popularity,
                                           track_release_date)),
                                  columns=columns,)

df_artist_top_tracks['ReleaseDate'] = pd.to_datetime(df_artist_top_tracks['ReleaseDate'],
                                                     format='%Y-%m-%d',
                                                     errors='ignore')


# print(results['tracks'][0])
print(df_artist_top_tracks.info())
df_artist_top_tracks.head()
```

```python
# Artist's Top Tracks Funcition

def top_tracks(artist_id):
    
    '''
    
    INPUT: Individual artist_id
    OUTPUT: DataFrame 'SpotifyTrackID','ArtistsName', 'TrackName', 'TrackPopularity', 'ReleaseDate'
    
    '''


    results=spotify.artist_top_tracks(artist_id)

    top_track_author=[]
    artist_top_tracks=[]
    track_popularity=[]
    track_release_date=[]
    track_spotify_id=[]

    for track in results['tracks']:
        artist_top_tracks.append(track['name'])
        track_popularity.append(track['popularity'])
        track_release_date.append(track['album']['release_date'])
        track_spotify_id.append(track['id'])
        top_track_author.append(track['artists'][0]['name'])

    columns = ['SpotifyTrackID','ArtistsName', 'TrackName', 'TrackPopularity', 'ReleaseDate']

    df_artist_top_tracks=pd.DataFrame(list(zip(track_spotify_id,
                                               top_track_author,
                                               artist_top_tracks,
                                               track_popularity,
                                               track_release_date)),
                                      columns=columns,)

    df_artist_top_tracks['ReleaseDate'] = pd.to_datetime(df_artist_top_tracks['ReleaseDate'],
                                                         format='%Y-%m-%d',
                                                         errors='ignore')


    #print(df_artist_top_tracks.info())
    return df_artist_top_tracks
```

```python
# Audio Features

# THIS FUNCTION RECEIVES A LIST OF SPOTIFYTRACKID

results = spotify.audio_features(tracks=['4SH86PVrzO2YbuOGDxzqqs',
                                        '13ZJckaESmpl62Y6YnTd7g'])

track_id=[]
track_danceability=[]
track_energy=[]
track_loudness=[]
track_instrumentalness=[]
track_liveness=[]
track_valence=[]
track_tempo=[]


for track in results:
    track_id.append(track['id'])
    track_danceability.append(track['danceability'])
    track_energy.append(track['energy'])
    track_loudness.append(track['loudness'])
    track_instrumentalness.append(track['instrumentalness'])
    track_valence.append(track['valence'])
    track_tempo.append(track['tempo'])
    
    
columns=['SpotifyTrackID','Danceability','Energy','LoudnessdB','Instrumentalness', 'Valence', 'Tempo']
    
df_audio_features_top_tracks=pd.DataFrame(list(zip(track_id,
                                                   track_danceability,
                                                   track_energy,
                                                   track_loudness,
                                                   track_instrumentalness,
                                                   track_valence,
                                                   track_tempo)), columns=columns)

print(df_audio_features_top_tracks.info())
df_audio_features_top_tracks.head()
```

```python
# Audio Features by Tracks List Function

def audio_features_track(list_of_tracks):
    '''
    
    INPUT: LIST OF TRACKS
    OUTPUT: DataFrame 
    
    '''
    
    results = spotify.audio_features(tracks=list_of_tracks)

    track_id=[]
    track_danceability=[]
    track_energy=[]
    track_loudness=[]
    track_instrumentalness=[]
    track_liveness=[]
    track_valence=[]
    track_tempo=[]


    for track in results:
        track_id.append(track['id'])
        track_danceability.append(track['danceability'])
        track_energy.append(track['energy'])
        track_loudness.append(track['loudness'])
        track_instrumentalness.append(track['instrumentalness'])
        track_valence.append(track['valence'])
        track_tempo.append(track['tempo'])


    columns=['SpotifyTrackID','Danceability','Energy','LoudnessdB','Instrumentalness', 'Valence', 'Tempo']

    df_audio_features_top_tracks=pd.DataFrame(list(zip(track_id,
                                                       track_danceability,
                                                       track_energy,
                                                       track_loudness,
                                                       track_instrumentalness,
                                                       track_valence,
                                                       track_tempo)), columns=columns)

    # print(df_audio_features_top_tracks.info())
    return df_audio_features_top_tracks
```
